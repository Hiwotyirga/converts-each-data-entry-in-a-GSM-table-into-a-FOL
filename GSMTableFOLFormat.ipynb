```python
for key, value in gsm.metadata.items():
    print(f"{key}: {value[0]}")  
```

```python
gsm.metadata
```

```python
all_predicates = []

for key, value in gsm.metadata.items():
    metadata_value = value[0] 
    fol_predicate = f'has("{gsm_id}","{key}", "{metadata_value}").'
    all_predicates.append(fol_predicate)  

```

```python
all_predicates
```

```python
records = df.to_dict('records')
```

```python
records
```

```python
expression_predicates = []
for idx, record in enumerate(records, start=1):  # idx = row number
    gene_id = record["ID_REF"]
    value = record["VALUE"]  
    expr_pred = f'expression_gene("{gsm_id}", "{gene_id}", {value}, row_{idx}).'
    expression_predicates.append(expr_pred)
```

```python
expression_predicates
```

```python
with open("Metadata.metta", "w") as f:
    for predicate in all_predicates: 
        f.write(predicate + "\n")    

with open("gene_expression.metta", "w") as f:
    for predicate in expression_predicates:
        f.write(predicate + "\n")
```
