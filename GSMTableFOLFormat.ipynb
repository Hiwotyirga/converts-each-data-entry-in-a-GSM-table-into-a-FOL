```python
import pandas
import GEOparse
```

```python
gse = GEOparse.get_GEO(filepath=".\GSE34378_family.soft.gz")
```

```python
gsm_id = list(gse.gsms.keys())[0]  
```

```python
gsm = gse.gsms[gsm_id]
```

```python
gsm.table.head(5)
```

```python
gsm.table.columns
```

```python
for key, value in gsm.metadata.items():
    print(f"{key}: {value[0]}")  
```

```python
all_predicates = []


for key, value in gsm.metadata.items():
    metadata_value = value[0] if isinstance(value, (list, tuple)) else value
    clean_key = key.lower().replace(" ", "_").replace("/", "_")
    fol_predicate = f'has_{clean_key}("{gsm_id}", "{metadata_value}").'
    all_predicates.append(fol_predicate)  

```

```python
all_predicates
```

```python
expression_predicates = []  

if hasattr(gsm, "table"):
    print("\nExpression Predicates:")
    df = gsm.table.T  
    for sample_row in df.head(5).itertuples():
        sample_id = sample_row.Index
        for i, (gene, expr) in enumerate(zip(df.columns, sample_row[1:])):
            try:
                expr_float = float(expr)
                expr_pred = f'expression_gene("{sample_id}", "{gene}", {expr_float:.2f}).'
                expression_predicates.append(expr_pred)
                
                if expr_float > 5.0:
                    high_pred = f'high_expression("{sample_id}", "{gene}").'
                    expression_predicates.append(high_pred)
                    
                print(expr_pred)  
                if expr_float > 5.0:
                    print(high_pred)
            except ValueError:
                continue
```

```python
with open(" GSM_table_(Metadata)+into_FOL_predicate_format.txt", "w") as f:
    for predicate in all_predicates: 
        f.write(predicate + "\n")    

with open(" GSM_table_(expression_predicates)+into_FOL_predicate_format.txt", "w") as f:
    for predicate in expression_predicates:
        f.write(predicate + "\n")
```
